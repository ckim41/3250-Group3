#### Data cleaning and normalization

```python
#remove all special values (unknown to us) in date-time columns, prepare for using date series as index.
#make a copy "df" for further analysis, avoid mess up the original data "accident".

df = accident[:]
df[['c_mnth','c_wday','c_hour','c_rcfg','c_wthr','c_rsur','c_traf']]=df[['c_mnth','c_wday','c_hour','c_rcfg','c_wthr','c_rsur','c_traf']].replace('[^0-9]+',np.nan,regex=True)

df.dropna(axis=0,subset=['c_mnth','c_wday','c_hour','c_rcfg','c_wthr','c_rsur','c_traf'],inplace=True)
df
```

```python
#generate the date-time column "date", and assign it as the index.

#df['date'] = pd.DatetimeIndex(df['c_year'].map(str) + '-' + df['c_mnth'] + '-' + df['c_wday'] + ' ' + df['c_hour'] + ':0:0')
df['date'] = pd.PeriodIndex(df['c_year'].map(str) + '-' + df['c_mnth'], freq='M')
df
```

```python
# this line use 'date' as index, meke 'df' becoming time series.

df=df.set_index('date')
df
```

#### (1) Trend over the whole period and seasonality identification

```python
df['fatal']=np.where(df['c_sev']==1,1,0)
df['non_fatal']=np.where(df['c_sev']==2,1,0)
```

```python
df_1=df.groupby('date')['fatal','non_fatal'].sum()
df_1
```

```python
#The overall diagram
plot1=df_1.plot(figsize=(15,5),title='Collision overall statistics')
plot1.set_xlabel("Date")
plot1.set_ylabel("Number of collisions")
```

```python
# Since the dateset period is 15 year, it's diffucult to identify a seasonality pattern in this scale
# Thus, a smaller period was selected to identify on-peak and off-peak of collisions
plot1=df_1[-60:].plot(figsize=(15,5),title='Collision overall statistics (recent 5 year)')
plot1.set_xlabel("Date")
plot1.set_ylabel("Number of collisions")
```

```python
# Due to a huge difference in number of fatal and non-fatal collisions, fatal collisions trend was analyzed in its own scale.
# Overall statistics
plot2=df_1['fatal'].plot(figsize=(15,5),title='Fatal collision statistics')
plot2.set_xlabel("Date")
plot2.set_ylabel("Number of collisions")
```

```python
# Fatal collision statistics for the recent 5 years
plot1=df_1['fatal'][-60:].plot(figsize=(15,5),title='Fatal collision statistics (recent 4 year)')
plot1.set_xlabel("Date")
plot1.set_ylabel("Number of collisions")
```

##### Conslusion: The  collision trend declines slightly over years and has pronounced seasonality pattern. The collision peak is in the middle of year and downturn is in Q1.

#### (2) Correlation between total number of collisions and fatal collisions
##### Autocorrelation plot demostrates an obvious seasonality with period of 12 months.

```python
# To confirm the seasonality assumtion, the autocorrelation of fatal and non-fatal collisions was performed for 
# the recent 5 years.
plt.figure(figsize=(15,5))
for c in df_1.columns:
    autocorrelation_plot(df_1[c][-60:],label=c)
```

##### The fatal and non-fatal collisions have strong positive correlation.

```python
sns.regplot('fatal', 'non_fatal', data=df_1)
```

#### The  correlation coefficient between fatal and non fatal collisions is 0.87

```python
corr = df_1.corr()
corr
```

#### 3.Analysis of collision risk dependancy on different week day and day hours

#### 4.Analysis of the following factors to collision rate:
####    a) Road configuration
####    b) Weather
####    c) Road surface
####    d) Traffic control

```python
#Number of collisions grouped by road configuration in % (Top 5)
rcfg_s1=df.groupby('c_rcfg')['c_sev'].count().sort_values(ascending=True)
total_collisions=rcfg_s1.sum()
rcfg_s2=rcfg_s1*100/total_collisions
df_2=pd.concat([rcfg_s1, rcfg_s2], axis=1)
df_2.columns=['sum','%']
df_2=df_2[-5:]
df_2
```

```python
df_2.index=['Railroad level crossing','Bridge','Intersection with parking lot entrance/exit',
            'Non-intersection','At an intersection']
ax=df_2['%'].plot.barh(color='slateblue',figsize=(15,5),title='Collisions by road configuration')
ax.xaxis.set_major_formatter(PercentFormatter())
```

```python
#Number of collisions grouped by weather condition in % (Top 5)
c_wthr_s1=df.groupby('c_wthr')['c_sev'].count().sort_values(ascending=True)
total_collisions=c_wthr_s1.sum()
c_wthr_s2=c_wthr_s1*100/total_collisions
df_3=pd.concat([c_wthr_s1, c_wthr_s2], axis=1)
df_3.columns=['sum','%']
df_3=df_3[-5:]
df_3
```

```python
df_3.index=['Fog, smog, mist','Snowing','Overcast, cloudy but no precipitation','Raining','Clear and sunny']
ax=df_3['%'].plot.barh(color='gold',figsize=(15,5),title='Collisions by weather condition')
ax.xaxis.set_major_formatter(PercentFormatter())
```

```python
#Number of collisions grouped by road surface in % (Top 5)
c_rsur_s1=df.groupby('c_rsur')['c_sev'].count().sort_values(ascending=True)
total_collisions=c_rsur_s1.sum()
c_rsur_s2=c_rsur_s1*100/total_collisions
df_4=pd.concat([c_rsur_s1, c_rsur_s2], axis=1)
df_4.columns=['sum','%']
df_4=df_4[-5:]
df_4
```

```python
df_4.index=['Slush, wet snow','Snow (fresh, loose snow)','Icy','Wet','Dry, normal']
ax=df_4['%'].plot.barh(color='darkmagenta',figsize=(15,5),title='Collisions by Road Surface')
ax.xaxis.set_major_formatter(PercentFormatter())
```

```python
#Number of collisions grouped by traffic control
c_traf_s1=df.groupby('c_traf')['c_sev'].count().sort_values(ascending=True)
total_collisions=c_traf_s1.sum()
c_traf_s2=c_traf_s1*100/total_collisions
df_5=pd.concat([c_traf_s1, c_traf_s2], axis=1)
df_5.columns=['sum','%']
df_5=df_5[-5:]
df_5
```

```python
df_5.index=['Pedestrian crosswalk','Yield sign','Stop sign',
            'Traffic signals fully operational','No control present']
ax=df_5['%'].plot.barh(color='navy',figsize=(15,5),title='Collisions by traffic control')
ax.xaxis.set_major_formatter(PercentFormatter())
```

##### Conclusion: The majority of collisions occur at intersection in a clean and sunny day when road surface is dry and no control signs are present. 
##### Based on other categories with significant weigth in the collision statistics, we can recommend to be more careful when you drive on wet or icy road and pay attentiomn near stop sign.
